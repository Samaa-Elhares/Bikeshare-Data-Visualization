# Basic Cleaning
   There are some issues need to be cleaned<br>
   * missing vlaues (start_station_id and name, end_station_id and name, member_birth_year and gender)
   * Wrong format in :
       * start_station_id (not object)
       * end_station_id (not_object)
       * bike_id (not_object)
       * start and end times (not timestamp)
       * member_birth_year(should be integer)

### Dropping missing values

```python
ford.dropna(inplace = True)
```

### Changing columns formats

```python
ford['start_time'] = pd.to_datetime(ford['start_time'])
ford['end_time'] = pd.to_datetime(ford['end_time'])
```

```python
ford['start_station_id'] = ford['start_station_id'].astype(str)
ford['end_station_id'] = ford['end_station_id'].astype(str)
ford['bike_id'] = ford['bike_id'].astype(str)
```

```python
ford.user_type.value_counts()
```

```python
ford.member_gender.value_counts()
```

```python
ford.bike_share_for_all_trip.value_counts()
```

```python
# Changing the three above columns into category type
ford.user_type = ford.user_type.astype('category')
ford.member_gender = ford.member_gender.astype('category')
ford.bike_share_for_all_trip = ford.bike_share_for_all_trip.astype('category')
```

```python
# getting age of a member
ford['age'] = 2021 - ford['member_birth_year']
ford['age'] = ford['age'].astype(int)
ford['member_birth_year'] = ford['member_birth_year'].astype(int)
```

```python
# Extracting month, dayofweek, hour, and date out of start_time column 
ford['start_month'] = ford['start_time'].dt.strftime('%B')
ford['dayofweek_name'] = ford['start_time'].dt.day_name()
ford['start_time_hour'] = ford['start_time'].dt.hour
ford['start_date'] = ford['start_time'].dt.date
```

```python
# end time hour
ford['end_time_hour'] = ford['end_time'].dt.hour
```

```python
ford.head()
```

```python
# Changing duration to miutes
ford['duration_minu'] = ford['duration_sec']/60
ford['duration_minu'] = ford['duration_minu'].astype(int)
```

```python
# Dropping not needed columns
```

## Univariate Exploration

> In this section, investigate distributions of individual variables. If

```python
base_color = sb.color_palette()[0]
```

## What are the most days for a bike ride?

```python
# Trips over the weekday

weekdays = ['Monday','Tuesday','Wednesday','Thursday','Friday',
            'Saturday','Sunday']

ax = sb.catplot(data = ford, x= 'dayofweek_name', kind = 'count',
                color = base_color, order =weekdays )
plt.xlabel('Week days')
plt.ylabel('Number of Trips')
plt.title('Trips Distribution over the Week')
plt.xticks(rotation = 15);
ax.fig.set_size_inches(10, 8);
```

> **Obs1:** It seems that days from Monday : Friday is the most frequent days for bikes trips but Thursdays are the most frquent,<br> Saturdays and Sundays aren't very popular as the rest.

## What time do people go for a bike trip?

```python
# trip by hour
ax = sb.catplot(data = ford, x = 'start_time_hour', kind = 'count',
          color = base_color)
plt.xlabel('Hours per day')
plt.ylabel('Number of bike rides')
plt.title('Trips Distribution over the day')
plt.xticks(rotation = 15)
ax.fig.set_size_inches(10, 8);
```

> **Obs2:** The most usage is during 7:9 am and 4:7 pm

## During Which month do trips take place ? 

```python
# Trips over month

month = ['January','February','March','April','May','June','July',
        'August','September','October','November','December']
ax = sb.catplot(data = ford, x= 'start_month', kind = 'count',
                color = base_color, order =month )
plt.xlabel('Month')
plt.ylabel('Number of Trips')
plt.title('Trips Distribution over the Month')
plt.xticks(rotation = 90)
ax.fig.set_size_inches(10, 8);
```

> **Obs3:** Apperently There are no trips over the year but in February.

## Is there a gender difference in trip duration?

```python
plt.figure(figsize=[10,8])
sb.boxplot(data = ford, x='member_gender', y='duration_minu',
           color= base_color)
plt.xlabel('Gender')
plt.ylabel('Trip duration in minutes')
plt.title('Gender and Trip Duration')
plt.ylim(0,80);
```

> **Obs8:** Males had shorter trip duration, but the difference isn't significant, a lower median and IQR.

## Is there a User Type difference in trip duration?

```python
plt.figure(figsize=[10,8])
sb.boxplot(data = ford, x='user_type', y = 'duration_minu',
             color = base_color)
plt.xlabel('User Type')
plt.ylabel('Trip duration in minutes')
plt.title('User Type and Trip Duration')
plt.ylim(0,80);
```

> **Obs9:** Subscribers have shorter trips

## Weekdays and Trip Duration

```python
plt.figure(figsize=[10,8])
weekdays_categ = pd.api.types.CategoricalDtype( ordered = True,categories=weekdays)
ford['dayofweek_name'] = ford['dayofweek_name'].astype(weekdays_categ)
sb.boxplot(data = ford, x='dayofweek_name', y = 'duration_minu',
             color = base_color)
plt.xlabel('Weekdays')
plt.ylabel('Trip duration in minutes')
plt.title('Weekdays and Trip Duration')
plt.ylim(0,80)
plt.xticks(rotation = 15);
```

> **Obs10:** Saturdays and Sundays have the highest trips duration ( It's the weekend)

## Who rented bikes more during the week, Customers or Subscribers?

```python
plt.figure(figsize=[10,8])
ford_user_week = ford.groupby(['dayofweek_name','user_type']).size().reset_index()
sb.pointplot(data= ford_user_week, x='dayofweek_name',y = 0,
              hue = 'user_type')
plt.xlabel('Weekday')
plt.ylabel('Count of Trips')
plt.title('User Type renting over Weekdays')
plt.xticks(rotation=15);
```

> **Obs11:** Subscribers are killin it over all days<br>
noticing the decrease duting weekend days they are propably using it to go to work.

## consumption on a daily usage 

```python
plt.figure(figsize=[10,8])
sb.countplot(data = ford, x = 'start_time_hour', hue='user_type')
plt.xlabel('Hour per day')
plt.ylabel('Count')
plt.title('User Type Frequency over Weekdays');
```

> **Obs12:** Subscribers uses during the morning to go to work and by the late afternoon.

## Does age affect Trip Duration?

```python
plt.figure(figsize=[10,8])
ford['duration_minu']
plt.scatter(data = ford, x= 'duration_minu', y ='age', alpha = 5/20)
plt.title('Age effect on trip duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Age frquency')
plt.ylim(20,60);
```

## What the average trip duration through weekdays for Subscribers and Customers?

```python
plt.figure(figsize=[10,8])
sb.pointplot(data=ford, x='dayofweek_name', y='duration_minu',
             hue = 'user_type', dodge=.2)
plt.xlabel('Weekdays')
plt.ylabel('Trip Duration Avg (min)')
plt.title('Trip duration Avg for users through the week')
plt.xticks(rotation=15);
```

## What the average trip duration through Day-hours for Subscribers and Customers?

```python
plt.figure(figsize=[10,8])
sb.barplot(data=ford,x='start_time_hour', y='duration_minu',ci = None,
          hue='user_type')
plt.xlabel('Hours')
plt.ylabel('Trip Duration Avg (min)')
plt.title('Trip duration Avg for users through the Day-Hours');
```

> **Obs16:** Subscribers ride shorter compared to customers during the weekdays, also their trip duration is almost steady the whole day.
